```python
amz=pd.read_csv('Amazon- Movies and Films.csv')
```

```python
amz.rename(columns={'Unnamed: 0':'Index'},inplace=True)
```

```python
amz.set_index('Index',inplace=True)
```

```python
amz['Format'].value_counts()
```

```python
amz.isnull().sum()
```

```python
amz['ReleaseYear'] = amz['ReleaseYear'].fillna(amz['ReleaseYear'].mode()[0])
amz['MPAA_Rating'] = amz['MPAA_Rating'].fillna(amz['MPAA_Rating'].mode()[0])
amz['Price'] = amz['Price'].fillna(amz['Price'].mean())
```

```python
amz.isnull().sum()
```
